```python
import GetOldTweets3 as got
from bs4 import BeautifulSoup
import datetime
import time
```

```python
# 가져올 범위를 정의
days_range = []

start = datetime.datetime.strptime("2020-05-10", "%Y-%m-%d")
end = datetime.datetime.strptime("2020-05-21", "%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

for date in date_generated:
    days_range.append(date.strftime("%Y-%m-%d"))
    
# 수집 기간 맞추기
start_date = days_range[0]
end_date = (datetime.datetime.strptime(days_range[-1], "%Y-%m-%d") 
            + datetime.timedelta(days=1)).strftime("%Y-%m-%d") # setUntil이 끝을 포함하지 않으므로, day + 1

print("=== 설정된 트윗 수집 기간은 {} 에서 {} 까지 입니다 ===".format(days_range[0], days_range[-1]))
print("=== 총 {}일 간의 데이터 수집 ===".format(len(days_range)))
```

```python
# 트윗 수집 기준 정의
tweetCriteria = got.manager.TweetCriteria().setQuerySearch('iot security')\
                                           .setSince(start_date)\
                                           .setUntil(end_date)\
                                           .setMaxTweets(-1)

# 수집
print("Collecting data start.. from {} to {}".format(days_range[0], days_range[-1]))
start_time = time.time()

tweet = got.manager.TweetManager.getTweets(tweetCriteria)

print("Collecting data end.. {0:0.2f} Minutes".format((time.time() - start_time)/60))
print("=== Total num of tweets is {} ===".format(len(tweet)))
```

```python
# 원하는 변수 골라서 저장하기

from random import uniform
# from tqdm.notebook import tqdm

# initialize
tweet_list = []
start_time = time.time()

for index in tweet:
    
    # 메타데이터 목록 
    username = index.username 
    content = index.text
    tweet_date = index.date.strftime("%Y-%m-%d")
     
    # 결과 합치기
    info_list = [tweet_date, username, content]
    tweet_list.append(info_list)
    
    # 휴식 
    time.sleep(uniform(1,2))
    
print("Processing data end.. {0:0.2f} Minutes".format((time.time() - start_time)/60))
```

```python
#저장
import pandas as pd

twitter_df = pd.DataFrame(tweet_list, columns = ["date", "user_name", "text"])

# csv 파일 만들기
twitter_df.to_csv("twitter_data_{}_to_{}.csv".format(days_range[0], days_range[-1]), index=False)
print("=== {} tweets are successfully saved ===".format(len(tweet_list)))
```
