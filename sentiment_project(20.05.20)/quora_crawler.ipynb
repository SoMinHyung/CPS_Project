```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
import csv
```

```python


class Crawler():

    def __init__(self):
        print("크롤링 시작")

    def make_csv(self):
        file = open("Quora.csv", "w", newline='', encoding  = 'UTF-8')
        wr = csv.writer(file)
        wr.writerow(["comments"])
        file.close()

    def write_csv(self, comments):
        file = open("Quora.csv", "a", newline='', encoding  = 'UTF-8')
        wr = csv.writer(file)
        for comment in comments :
            wr.writerow([comment])
        file.close()  

    def comment_scrap(self, driver, question):
        comments_list =[]
        
        driver.get("https://www.quora.com" + question)
        driver.implicitly_wait(10)
        
        for num in range(3):            
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)

        html = driver.page_source
        bs = BeautifulSoup(html, "html.parser") 
        

        #댓글 수집
        comments_boxs = bs.find_all("p", class_="q-text qu-display--block")
        for comments_box in comments_boxs:
            comment = ""
            for comments in comments_box.select('span'):
                comment += comments.text
            comments_list.append(comment)
        print(str(len(comments_list))+"개")
        self.write_csv(comments_list)
        time.sleep(30)
        
        driver.implicitly_wait(10)
        
        
    def scrap(self, num):
        path = os.getcwd()+"\chromedriver.exe"
        driver = webdriver.Chrome(path)

        try:
            self.make_csv()
            driver.get("https://www.quora.com")
            driver.implicitly_wait(10)

            email = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/div[1]/div/div[2]/div/div/div/form/div[2]/div[1]/input')
            driver.execute_script("arguments[0].setAttribute('value', 'handsomebuda@gmail.com')", email)

            pw = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/div[1]/div/div[2]/div/div/div/form/div[2]/div[2]/input')
            driver.execute_script("arguments[0].setAttribute('value', 'Buda0970!')", pw)

            login = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/div[1]/div/div[2]/div/div/div/form/div[2]/div[3]/input')
            driver.execute_script("arguments[0].click();", login)
            driver.implicitly_wait(10)
            
            driver.get("https://www.quora.com/search?q=iot%20security")
            time.sleep(5)
            
            numPage = 4
            
            for num in range(numPage):            
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(3)
            
            count = 0
            html = driver.page_source
            bs = BeautifulSoup(html, "html.parser")
            questions = bs.find("div", class_="PagedList QueryResultsList").find_all("a", class_="question_link")
                    
            for question in questions:                
                #n번째 질문 페이지로 이동 후 댓글 크롤링, csv 파일에 저장
                self.comment_scrap(driver, question.attrs['href'])

                count += 1
                print(str(count) + "번 째 질문 완료")

            time.sleep(5)
            print("크롤링 종료")

        finally:
            time.sleep(3)
            driver.quit()

```

```python
c = Crawler()
c.scrap(1)
```
