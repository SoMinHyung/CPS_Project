```python
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
```

```python
path = os.getcwd()+"\chromedriver.exe"
driver = webdriver.Chrome(path)
```

```python
"""
이 크롤러의 목적


네이버 증권에서 당일 거래량 1위 종목을 찾아서, 그 종목의 종목토론방으로 가보자.


그리고 사람들이 그 종목에 대해서 어떤 글을 작성하고 있는지 제목을 알아보도록 하자

"""
```

```python
try:
    driver.get("https://finance.naver.com/")
    time.sleep(1)
    
    #검색할 검색어는 거래량 상위 1등 종목
    html = driver.page_source
    bs = BeautifulSoup(html, "html.parser")
    searchIndex = bs.find("tbody", id="_topItems1").find_all("a")[0].text
    
    #종목을 검색
    element = driver.find_element_by_xpath('//*[@id="stock_items"]')
    element.send_keys(searchIndex)
    driver.find_element_by_class_name("snb_search_btn").click()
    
    time.sleep(1)
    
    #해당 종목의 종목토론실로 이동
    driver.find_element_by_class_name("tab7").click()
    
    title = []
    
    #최신 3페이지의 데이터만을 받아오자.
    for i in range(5):
        
        time.sleep(2)
        
        html = driver.page_source
        bs = BeautifulSoup(html, "html.parser")
        
        title.append("page"+ str(i+1))
    
        #한 페이지의 제목 목록을 받아옴
        contents = bs.find("table", class_="type2").find_all("td", class_="title")
        
        for c in contents:
            #각 컨텐츠에서 div태그의 title클래스 데이터를 가져와서, 그 내부의 strong태그의 text만을 반환
            title.append(c.find("a")['title'])
        
        #다음 페이지로 이동
        if i == 0:
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/table[2]/tbody/tr/td[2]/table/tbody/tr/td[' + str(i+2) +']/a').click()
        else:
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/table[2]/tbody/tr/td[2]/table/tbody/tr/td[' + str(i+3) +']/a').click()
    
finally:
    
    for t in title:
        if t.find("page") != -1 :
            print()
            print(t)
        else:
            print(t)
    
    time.sleep(3)
    driver.quit()
```
